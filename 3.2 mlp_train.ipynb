```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('250google_vector.csv')
df.head()
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
```

```python
X=df.iloc[:,0:300].values.tolist()
y=df.iloc[:,300].values.tolist()
```

```python
layer_sizes = range(1,101)
mlp_scores = []
max=0
max_number=0
for layer_number in layer_sizes:
    clf = MLPClassifier(solver='lbfgs', alpha=1e-5,
                     hidden_layer_sizes=(layer_number), random_state=1)
    scores = cross_val_score(clf,X,y,cv=10,scoring='accuracy')
    mlp_scores.append(scores.mean())
    if scores.mean() > max:
        max=scores.mean()
        max_number = layer_number
print(max)
print(max_number)
```

```python
plt.plot(layer_sizes,mlp_scores)
plt.xlabel('Value of hidden layer for MLP')
plt.ylabel('Cross-Validated Accuracy')
plt.show()
```
